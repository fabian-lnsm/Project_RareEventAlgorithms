```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Define the modified double-well potential
def potential(x, t, mu):
    return 0.25 * x**4 - 0.5 * x**2 - mu * x * t

#set coupling parameter
mu = 0.001

# Plotting the potential for different points
x = np.linspace(-2, 2, 1000)
t = np.linspace(-100, 100, 10)

minima_left = np.zeros((len(t),2))
minima_right = np.zeros((len(t),2))
fig, ax = plt.subplots(dpi=250)
for i in range(len(t)):
    y_values = potential(x, t[i], mu)
    min_left = np.argmin(y_values[x < 0])
    minima_left[i, 0] = t[i]
    minima_left[i, 1] = min_left
    min_right = np.argmin(y_values[x > 0]) + len(x[x < 0])
    minima_right[i, 0] = t[i]
    minima_right[i, 1] = min_right
    ax.plot(x, potential(x, t[i], mu), label='t = ' + str(t[i]))
ax.set_xlabel('x')
ax.set_ylabel('V(x,t)')
ax.grid(True)
#ax.legend()

fig,ax = plt.subplots(dpi=250)
ax.plot(minima_left[:,0], minima_left[:,1], label='left minimum')

```
